```python
import pandas as pd
from sqlalchemy import create_engine
import requests
import json
from pprint import pprint
```

```python
url = "http://www.omdbapi.com/?t="
api_key = "&apikey=77d3b433"
```

# EXTRACT

### Store CSV into DataFrame

```python
csv_file = "Resources/netflix_titles.csv"
netflix_df = pd.read_csv(csv_file)
netflix_df
```

# TRANSFORM

### Create new data with select columns

```python
#Filter netflix titles to Movies only
netflix_movies = netflix_df.loc[netflix_df['type'] == 'Movie']
# netflix_movies = netflix_movies.loc[netflix_movies['country'] == 'United States']
netflix_movies = netflix_movies[netflix_movies['listed_in']=='Documentaries']
netflix_movies
```

```python
#pull only select columns from the data
new_netflix_df = netflix_movies[['show_id','title', 'country', 'release_year', 'rating']].copy()
#rename id column
netflix_transformed = new_netflix_df.rename(columns={"show_id": "id"})
netflix_transformed
```

### API OMBDB

```python
imdb_rating = []
meta_rating = []
awards =[]
not_found = 0
found_movies = []
movies = netflix_transformed['title']

for movie in movies:
    response = requests.get (url + movie + api_key)
    omdb = response.json()
    try:
        imdb_rating.append(omdb["imdbRating"])
        meta_rating.append(omdb["Metascore"])
        awards.append(omdb["Awards"])
        found_movies.append(movie)
    except:
        not_found = not_found + 1
        pass
```

```python
#Create dataframe from information acquired from API
omdb_df = pd.DataFrame(list(zip(found_movies, imdb_rating, meta_rating, awards)),columns = ["Movie_Title", "IMDB_Rating", "Metacritic_Rating", "Awards"])
omdb_df = omdb_df[omdb_df!="N/A"].dropna()
omdb_df
```

### Connect to local database

```python
rds_connection_string = "postgres:viperson1996@localhost:5432/netflix_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```
